```python
class Node:
    def __init__(self,data):
        self.data=data
        
class Single_LinkedList:
    def __init__(self):
        self.head=None
        
        
    def traversal(self):
        n=self.head
        
        if n is None:
            print("Linked List is Empty")
        else:
            while n is not None:
                print(n.data,end="--->")
                n=n.next
                
    def add_begin(self,data):
        newnode=Node(data)
        newnode.next=self.head
        self.head=newnode
        
    def add_ending(self,data):
        newnode=Node(data)
        n=self.head
        while n.next is not None:
            n=n.next
        n.next=newnode
        newnode.next=None
        
    def add_middle_after(self,next_node,data):
        newnode=Node(data)
        n=self.head
        
        while n is not None:
            if n.data==next_node:
                break
            n=n.next
        if n is None:
            print("Node is not present in LL")
        else:
            newnode.next=n.next
            n.next=newnode
            
    def add_middle_before(self,next_node,data):
        newnode=Node(data)
        n=self.head
        
        while n.next is not None:
            if n.next.data==next_node:
                break
            n=n.next
        if n is None:
            print("Node is not present in LL")
        else:
            newnode.next=n.next
            n.next=newnode
            
    def delete_begin(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            n=self.head
            self.head=n.next
            
    def delete_middle(self,x):
        if self.head is None:
            print("Linked List is Empty")
        else:
            n=self.head
            while n.next is not None:
                if x=n.next.data:
                    break
                n=n.next
        print("current ele",n.data)

            
    def delete_end(self):
        n=self.head
        if self.head is None:
            print("Linked List is empty")
        else:
            while n.next.next is not None:
                n=n.next
            n.next=None
            

        
s=Single_LinkedList()
s.add_begin(10)
s.add_begin(20)
s.add_begin(30)
s.add_ending(100)
s.add_middle_after(20,200)
s.add_middle_before(10,300)
s.delete_begin()
s.delete_end()
s.delete_middle(300)
s.traversal()
```
