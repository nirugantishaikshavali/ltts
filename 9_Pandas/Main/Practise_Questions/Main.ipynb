```python
res=pd.merge(movies_df,revenue_df,on="movie_id")
res
```

```python
res['profit'] = res['revenue'] - res['budget']
res
```

```python
res['release_year'] = pd.to_datetime(res['release_date']).dt.year
res
```

```python
top_movies_by_year = res.groupby('release_year').apply(lambda x:x.nlargest(2,"profit")).reset_index(drop=True)
```

```python
print("Top 2 movies by year based on profit:")
print(top_movies_by_year[['release_year', 'movie_name', 'genre', 'profit']])
```
